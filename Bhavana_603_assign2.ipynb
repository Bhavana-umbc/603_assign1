```python
#My DoB is MAy 29 2002
#so according to the instructions in the assignment the book I have selected is Harry Potter and the Order of the Phoenix
#for file 1 i have take 10 pages after 29
#fro file 2 i have taken 10 pages after 102
from collections import defaultdict
import re
from spellchecker import SpellChecker

#spliting the text to words
def tokenize(text):
    text = text.lower()
    words = re.findall(r'\b\w+\b', text)
    return words

#Map reduce function for word count
def map_reduce_word_count(filename):
    word_counts = defaultdict(int)
    with open(filename, 'r', encoding='utf-8') as file:
        for line in file:
            words = tokenize(line)
            for word in words:
                word_counts[word] += 1
    return word_counts

#Map Reduce for file 2 for non english words count
#And I have used spell checker for non englist word lookup
def map_reduce_non_english_count(filename):
    spell = SpellChecker()
    non_english_counts = defaultdict(int)
    with open(filename, 'r', encoding='utf-8') as file:
        for line in file:
            words = tokenize(line)
            misspelled = spell.unknown(words)
            for word in misspelled:
                non_english_counts[word] += 1
    return non_english_counts

#i am using google collab so we are adding the file path
file1_path = '/content/file1.txt'
file2_path = '/content/file2.txt'

#running the tasks
word_counts_file1 = map_reduce_word_count(file1_path)
non_english_counts_file2 = map_reduce_non_english_count(file2_path)

print("Occurrences of each word in the file 1:")
for word, count in sorted(word_counts_file1.items(), key=lambda x: x[1], reverse=True):
    print(f'{word}: {count}')

print("\nOccurance of non-English words in file2.txt:")
for word, count in sorted(non_english_counts_file2.items(), key=lambda x: x[1], reverse=True):
    print(f'{word}: {count}')
```
